```python
import matplotlib.pyplot as plt
import numpy as np
dataset = 'Heart Disease'
x = np.linspace(0, 0.5, 1000)
app1 = [0.2]
app2 = [0.18]

plt.scatter(app1, app2, c='black', label = 'Lasso Regression')

app1 = [0.32]
app2 = [0.28]

plt.scatter(app1, app2, c='blue',label = 'Linear Regression')


app1 = [0.29]
app2 = [0.2]

plt.scatter(app1, app2, c='green',label = 'Random Forest Regression')

app1 = [0.4]
app2 = [0.28]

plt.scatter(app1, app2, c='orange', label = 'Decision Tree Regression')

app1 = [0.33]
app2 = [0.14]

plt.scatter(app1, app2, c='red',label = 'AdaBoost Regression')
plt.plot(x, x + 0, linestyle='dotted')
string = 'MAE of initial data vs recent data of test person:'+ dataset
plt.title(string)
plt.xlabel('Initial Data')
plt.ylabel('Recent Data')
plt.legend(loc='best')
plt.savefig(string)
plt.show()
```
