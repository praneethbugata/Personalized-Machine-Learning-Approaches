```python
import matplotlib.pyplot as plt
import numpy as np
dataset = 'Heart Disease'
x = np.linspace(0, 1, 1000)
app1 = [0.75]
app2 = [0.89]

plt.scatter(app1, app2, c='black', label = 'Logistic Regression')

app1 = [0.28]
app2 = [0.24]

plt.scatter(app1, app2, c='blue',label = 'Support Vector Machine')

app1 = [0.63]
app2 = [0.8]

plt.scatter(app1, app2, c='orange', label = 'Decision Tree')

app1 = [0.82]
app2 = [0.82]

plt.scatter(app1, app2, c='green',label = 'Random Forest')

app1 = [0.47]
app2 = [0.69]

plt.scatter(app1, app2, c='red',label = 'AdaBoost')
plt.plot(x, x + 0, linestyle='dotted')
string = 'Similar Person AUC of proposed approaches:'+ dataset
plt.title(string)
plt.xlabel('Proposed Approach_1')
plt.ylabel('Proposed Approach_2')
plt.legend(loc='best')
plt.savefig(string)
plt.show()
```
